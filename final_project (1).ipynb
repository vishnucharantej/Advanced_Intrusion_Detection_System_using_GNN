```python
class MLPPredictor(nn.Module):
    def __init__(self, in_features, out_classes):
        super().__init__()
        self.W = nn.Linear(in_features * 2, out_classes)

    def apply_edges(self, edges):
        h_u = edges.src['h']
        h_v = edges.dst['h']
        global score
        global emb
        emb = th.cat([h_u, h_v], 1)
        score = self.W(th.cat([h_u, h_v], 1))
        return {'score': score}

    def forward(self, graph, h):
        with graph.local_scope():
            graph.ndata['h'] = h
            graph.apply_edges(self.apply_edges)
            return graph.edata['score']
```

```python
def compute_accuracy(pred, labels):
    return (pred.argmax(1) == labels).float().mean().item()
```

```python
class SAGELayer(nn.Module):
    def __init__(self, ndim_in, edims, ndim_out, activation):
        super(SAGELayer, self).__init__()
        ### force to outut fix dimensions
        self.W_msg = nn.Linear(ndim_in + edims, ndim_out)
        ### apply weight
        self.W_apply = nn.Linear(ndim_in + ndim_out, ndim_out)
        self.activation = activation

    def message_func(self, edges):
        return {'m': self.W_msg(th.cat([edges.src['h'], edges.data['h']], 2))}

    def forward(self, g_dgl, nfeats, efeats):
        with g_dgl.local_scope():
            g = g_dgl
            g.ndata['h'] = nfeats
            g.edata['h'] = efeats
            # Eq4
            g.update_all(self.message_func, fn.mean('m', 'h_neigh'))
            # Eq5
            g.ndata['h'] = F.relu(self.W_apply(th.cat([g.ndata['h'], g.ndata['h_neigh']], 2)))
            return g.ndata['h']


class SAGE(nn.Module):
    def __init__(self, ndim_in, ndim_out, edim, activation, dropout):
        super(SAGE, self).__init__()
        self.layers = nn.ModuleList()
        self.layers.append(SAGELayer(ndim_in, edim, 128, activation))
        self.layers.append(SAGELayer(128, edim, ndim_out, activation))
        self.dropout = nn.Dropout(p=dropout)

    def forward(self, g, nfeats, efeats):
        for i, layer in enumerate(self.layers):
            if i != 0:
                nfeats = self.dropout(nfeats)
            nfeats = layer(g, nfeats, efeats)
        return nfeats.sum(1)
```

```python
class Model(nn.Module):
    def __init__(self, ndim_in, ndim_out, edim, activation, dropout):
        super().__init__()
        self.gnn = SAGE(ndim_in, ndim_out, edim, activation, dropout)
        self.pred = MLPPredictor(ndim_out, 5)
    def forward(self, g, nfeats, efeats):
        h = self.gnn(g, nfeats, efeats)
        return self.pred(g, h)
```

```python
def plot_confusion_matrix(cm,
                          target_names,
                          title='Confusion matrix',
                          cmap=None,
                          normalize=True):

    import matplotlib.pyplot as plt
    import numpy as np
    import itertools

    accuracy = np.trace(cm) / float(np.sum(cm))
    misclass = 1 - accuracy

    if cmap is None:
        cmap = plt.get_cmap('Blues')

    plt.figure(figsize=(12, 12))
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()

    if target_names is not None:
        tick_marks = np.arange(len(target_names))
        plt.xticks(tick_marks, target_names, rotation=45)
        plt.yticks(tick_marks, target_names)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]


    thresh = cm.max() / 1.5 if normalize else cm.max() / 2
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        if normalize:
            plt.text(j, i, "{:0.4f}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")
        else:
            plt.text(j, i, "{:,}".format(cm[i, j]),
                     horizontalalignment="center",
                     color="white" if cm[i, j] > thresh else "black")


    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label\naccuracy={:0.4f}; misclass={:0.4f}'.format(accuracy, misclass))
    plt.show()
```

```python
def find_edges_by_attack_type(graph, edge_label):
    """
    Find example edges for each attack type in the graph
    """
    # Get all edges and their labels
    edge_indices = range(graph.number_of_edges())
    labels = edge_label.cpu().numpy()

    # Map class index to label name
    class_names = ["DDoS", "Reconnaissance", "Normal", "Theft"]

    # Find edges for each attack type
    example_edges = {}
    for class_idx, class_name in enumerate(class_names):
        # Find edges with this label
        edges_with_label = [i for i in edge_indices if labels[i] == class_idx]

        if edges_with_label:
            # Get the source and destination nodes for this edge
            src_nodes = graph.edges()[0][edges_with_label].cpu().numpy()
            dst_nodes = graph.edges()[1][edges_with_label].cpu().numpy()

            # Store example edges for this attack type
            example_edges[class_name] = [(src, dst) for src, dst in zip(src_nodes[:5], dst_nodes[:5])]

    return example_edges

def predict_attack(model, graph, node_features, edge_features, src_node, dst_node):
    """
    Predict attack type for a specific source-destination pair
    """
    # Set model to evaluation mode
    model.eval()

    # Get the edge ID
    edge_id = graph.edge_ids(src_node, dst_node)

    # Make prediction
    with th.no_grad():
        pred = model(graph, node_features, edge_features)
        edge_pred = pred[edge_id]

        # Handle different output formats
        if edge_pred.dim() > 1:
            # If prediction has batch dimension or extra dimensions
            edge_pred = edge_pred.squeeze()

        # Apply softmax on the correct dimension
        probabilities = F.softmax(edge_pred, dim=0)
        predicted_class = edge_pred.argmax().item()
        confidence = probabilities[predicted_class].item()

    # Map class index to label name
    class_names = ["DDoS", "Reconnaissance", "Normal", "Theft"]
    predicted_attack = class_names[predicted_class]

    return {
        'predicted_attack': predicted_attack,
        'confidence': confidence,
        'probabilities': {name: float(prob) for name, prob in zip(class_names, probabilities)}
    }

def predict_attack_demonstration(model, graph, node_features, edge_features, edge_label):
    """
    Interactive demonstration of attack prediction using valid edges
    """
    # First, find example edges for each attack type
    print("Finding example edges for each attack type...")
    example_edges = find_edges_by_attack_type(graph, edge_label)

    print("\nExample edges found in the graph:")
    for attack_type, edges in example_edges.items():
        print(f"\n{attack_type} connections:")
        for i, (src, dst) in enumerate(edges[:3], 1):  # Show up to 3 examples
            print(f"  {i}. Source: {src}, Destination: {dst}")

    print("\nEnter attack type to demonstrate prediction (or 'q' to quit)")

    while True:
        attack_choice = input("Choose attack type (DDoS, Reconnaissance, Normal, Theft, or 'q'): ")
        if attack_choice.lower() == 'q':
            break

        if attack_choice not in example_edges:
            print(f"Invalid choice. Available options: {', '.join(example_edges.keys())}")
            continue

        # Choose the first example edge for this attack type
        src_node, dst_node = example_edges[attack_choice][0]

        print(f"\nDemonstrating prediction for a {attack_choice} connection")
        print(f"Source node: {src_node}, Destination node: {dst_node}")

        try:
            # Make prediction
            result = predict_attack(model, graph, node_features, edge_features, src_node, dst_node)

            print(f"\nPrediction results:")
            print(f"Predicted attack type: {result['predicted_attack']}")
            print(f"Confidence: {result['confidence']:.2%}")

            print("\nProbabilities for each attack type:")
            for attack_type, prob in result['probabilities'].items():
                print(f"  {attack_type}: {prob:.2%}")

            if result['predicted_attack'] == attack_choice:
                print("✓ CORRECT PREDICTION")
            else:
                print("✗ INCORRECT PREDICTION")
            print()

        except Exception as e:
            print(f"Error making prediction: {e}")

    print("\nWould you like to predict an attack for a custom source-destination pair? (y/n)")
    if input().lower() == 'y':
        custom_prediction(model, graph, node_features, edge_features)

def custom_prediction(model, graph, node_features, edge_features):
    """
    Allow user to input custom source and destination nodes for prediction
    """
    print("\n--- CUSTOM ATTACK PREDICTION ---")
    print(f"Available node IDs range from 0 to {graph.number_of_nodes()-1}")

    while True:
        try:
            print("\nEnter source node ID (or 'q' to quit):")
            src_input = input()
            if src_input.lower() == 'q':
                break

            src_node = int(src_input)

            print("Enter destination node ID:")
            dst_node = int(input())

            # Validate node IDs
            if src_node < 0 or src_node >= graph.number_of_nodes() or dst_node < 0 or dst_node >= graph.number_of_nodes():
                print(f"Invalid node IDs. Must be between 0 and {graph.number_of_nodes()-1}")
                continue

            # Check if edge exists
            try:
                edge_id = graph.edge_ids(src_node, dst_node)
            except:
                print(f"No edge exists between node {src_node} and node {dst_node}")
                continue

            # Make prediction
            result = predict_attack(model, graph, node_features, edge_features, src_node, dst_node)

            print(f"\nPrediction results for {src_node} → {dst_node}:")
            print(f"Predicted attack type: {result['predicted_attack']}")
            print(f"Confidence: {result['confidence']:.2%}")

            print("\nProbabilities for each attack type:")
            for attack_type, prob in result['probabilities'].items():
                print(f"  {attack_type}: {prob:.2%}")

        except ValueError:
            print("Please enter valid integer node IDs")
        except Exception as e:
            print(f"Error: {e}")
```

```python
def find_edges_by_attack_type(graph, edge_label):
    """
    Find example edges for each attack type in the graph
    """
    # Get all edges and their labels
    edge_indices = range(graph.number_of_edges())
    labels = edge_label.cpu().numpy()

    # Map class index to label name
    class_names = ["DDoS", "Reconnaissance", "Normal", "Theft"]

    # Find edges for each attack type
    example_edges = {}
    for class_idx, class_name in enumerate(class_names):
        # Find edges with this label
        edges_with_label = [i for i in edge_indices if labels[i] == class_idx]

        if edges_with_label:
            # Get the source and destination nodes for this edge
            src_nodes = graph.edges()[0][edges_with_label].cpu().numpy()
            dst_nodes = graph.edges()[1][edges_with_label].cpu().numpy()

            # Store example edges for this attack type
            example_edges[class_name] = [(src, dst) for src, dst in zip(src_nodes[:5], dst_nodes[:5])]

    return example_edges

def predict_attack(model, graph, node_features, edge_features, src_node, dst_node):
    """
    Predict attack type for a specific source-destination pair
    """
    # Set model to evaluation mode
    model.eval()

    # Get the edge ID
    edge_id = graph.edge_ids(src_node, dst_node)

    # Make prediction
    with th.no_grad():
        pred = model(graph, node_features, edge_features)
        edge_pred = pred[edge_id]

        # Handle different output formats
        if edge_pred.dim() > 1:
            # If prediction has batch dimension or extra dimensions
            edge_pred = edge_pred.squeeze()

        # Apply softmax on the correct dimension
        probabilities = F.softmax(edge_pred, dim=0)
        predicted_class = edge_pred.argmax().item()
        confidence = probabilities[predicted_class].item()

    # Map class index to label name
    class_names = ["DDoS", "Reconnaissance", "Normal", "Theft"]
    predicted_attack = class_names[predicted_class]

    return {
        'predicted_attack': predicted_attack,
        'confidence': confidence,
        'probabilities': {name: float(prob) for name, prob in zip(class_names, probabilities)}
    }

def node_to_ip(node_id):
    """
    Convert a node ID to an IP address using a deterministic mapping
    """
    octet1 = 10
    octet2 = node_id % 255
    octet3 = (node_id // 255) % 255
    octet4 = (node_id // (255 * 255)) % 255 + 1  # Avoid 0

    return f"{octet1}.{octet2}.{octet3}.{octet4}"

def ip_to_node(ip_address):
    """
    Convert an IP address back to a node ID
    """
    try:
        octets = ip_address.split('.')
        if len(octets) != 4:
            return None

        # Extract the octets
        octet1 = int(octets[0])
        octet2 = int(octets[1])
        octet3 = int(octets[2])
        octet4 = int(octets[3])

        # Validate the IP format (basic check)
        if octet1 != 10 or octet4 < 1:
            return None

        # Reconstruct the node ID
        node_id = octet2 + (octet3 * 255) + ((octet4 - 1) * 255 * 255)
        return node_id
    except:
        return None

def predict_for_custom_ip_input(model, graph, node_features, edge_features):
    """
    Function to predict attack type for user-provided source and destination IPs
    """
    class_names = ["DDoS", "Reconnaissance", "Normal", "Theft"]

    # Show some sample IP addresses
    print("\nSample IP addresses in the network:")
    sample_nodes = [24537, 48964, 3432, 413, 5877]
    for node_id in sample_nodes:
        ip = node_to_ip(node_id)
        print(f"  Node {node_id}: {ip}")

    while True:
        try:
            print("\nEnter source IP address (or 'q' to quit):")
            src_input = input()
            if src_input.lower() == 'q':
                break

            # Convert IP to node ID
            src_node = ip_to_node(src_input)
            if src_node is None:
                print(f"Invalid IP format. Please use format like 10.x.y.z")
                continue

            print("Enter destination IP address:")
            dst_input = input()

            # Convert IP to node ID
            dst_node = ip_to_node(dst_input)
            if dst_node is None:
                print(f"Invalid IP format. Please use format like 10.x.y.z")
                continue

            # Validate node IDs
            if src_node < 0 or src_node >= graph.number_of_nodes() or dst_node < 0 or dst_node >= graph.number_of_nodes():
                print(f"Invalid IP addresses. Node IDs must be between 0 and {graph.number_of_nodes()-1}")
                continue

            # Make prediction
            result = predict_attack(model, graph, node_features, edge_features, src_node, dst_node)

            print(f"\nPrediction results for connection {src_input} → {dst_input}:")
            print(f"(Node {src_node} → Node {dst_node})")
            print(f"Predicted attack type: {result['predicted_attack']}")
            print(f"Confidence: {result['confidence']:.2%}")

            print("\nProbabilities for each attack type:")
            for attack_type, prob in result['probabilities'].items():
                print(f"  {attack_type}: {prob:.2%}")

        except Exception as e:
            print(f"Error making prediction: {e}")
            import traceback
            traceback.print_exc()

def automatic_predict_attack_demonstration(model, graph, node_features, edge_features, edge_label):
    """
    Automatic demonstration of attack prediction using edges from the graph
    """
    # First, find example edges for each attack type
    print("Finding example edges for each attack type...")
    example_edges = find_edges_by_attack_type(graph, edge_label)

    print("\nExample edges found in the graph:")
    for attack_type, edges in example_edges.items():
        print(f"\n{attack_type} connections:")
        for i, (src, dst) in enumerate(edges[:3], 1):  # Show up to 3 examples
            src_ip = node_to_ip(src)
            dst_ip = node_to_ip(dst)
            print(f"  {i}. Source: {src_ip} (Node {src}), Destination: {dst_ip} (Node {dst})")

    print("\n--- AUTOMATIC PREDICTION DEMONSTRATION ---")

    # Demonstrate prediction for one example of each attack type
    for attack_type, edges in example_edges.items():
        if not edges:
            print(f"\nNo examples found for {attack_type}")
            continue

        # Use the first example for this attack type
        src_node, dst_node = edges[0]
        src_ip = node_to_ip(src_node)
        dst_ip = node_to_ip(dst_node)

        print(f"\nDemonstrating prediction for a {attack_type} connection")
        print(f"Source: {src_ip} (Node {src_node}), Destination: {dst_ip} (Node {dst_node})")

        try:
            # Make prediction
            result = predict_attack(model, graph, node_features, edge_features, src_node, dst_node)

            print(f"\nPrediction results:")
            print(f"Predicted attack type: {result['predicted_attack']}")
            print(f"Confidence: {result['confidence']:.2%}")

            print("\nProbabilities for each attack type:")
            for attack_type_name, prob in result['probabilities'].items():
                print(f"  {attack_type_name}: {prob:.2%}")

            if result['predicted_attack'] == attack_type:
                print("✓ CORRECT PREDICTION")
            else:
                print("✗ INCORRECT PREDICTION")

        except Exception as e:
            print(f"Error making prediction: {e}")
            import traceback
            traceback.print_exc()

    # Demonstrate prediction for random edges from the graph
    print("\n--- RANDOM EDGE PREDICTIONS ---")
    print("Selecting 5 random edges from the graph for prediction...")

    # Get all edges
    all_edges = list(zip(graph.edges()[0].cpu().numpy(), graph.edges()[1].cpu().numpy()))

    # Randomly select up to 5 edges (or fewer if graph has fewer edges)
    import random
    num_samples = min(5, len(all_edges))
    random_edges = random.sample(all_edges, num_samples)

    for i, (src_node, dst_node) in enumerate(random_edges, 1):
        src_ip = node_to_ip(src_node)
        dst_ip = node_to_ip(dst_node)

        print(f"\nRandom Edge {i}: Source {src_ip} (Node {src_node}) → Destination {dst_ip} (Node {dst_node})")

        try:
            # Find the actual label
            edge_id = graph.edge_ids(src_node, dst_node)
            if th.is_tensor(edge_id):  # Using th instead of torch
                edge_id = edge_id.item()

            actual_label = edge_label[edge_id].item()
            class_names = ["DDoS", "Reconnaissance", "Normal", "Theft"]
            actual_attack_type = class_names[actual_label]

            print(f"Actual attack type: {actual_attack_type}")

            # Make prediction
            result = predict_attack(model, graph, node_features, edge_features, src_node, dst_node)

            print(f"Predicted attack type: {result['predicted_attack']}")
            print(f"Confidence: {result['confidence']:.2%}")

            if result['predicted_attack'] == actual_attack_type:
                print("✓ CORRECT PREDICTION")
            else:
                print("✗ INCORRECT PREDICTION")

        except Exception as e:
            print(f"Error making prediction: {e}")

    # Ask if user wants to try custom input prediction
    print("\nWould you like to predict attack types for custom source-destination IP pairs? (y/n)")
    choice = input().lower()
    if choice == 'y':
        predict_for_custom_ip_input(model, graph, node_features, edge_features)

    return example_edges

# Call the automatic demonstration function
example_edges = automatic_predict_attack_demonstration(model, G, G.ndata['h'], G.edata['h'], G.edata['label'])
```
